```python
# Select relevant columns
selected_columns = ['Main Meal','Rating Of  Selected Items of  Main Meal [Selected Food]']
selected_data = df[selected_columns]

# Extract unique 'Main Meal' values and their average ratings
unique_main_meal_ratings = selected_data.groupby('Main Meal')['Rating Of  Selected Items of  Main Meal [Selected Food]'].mean()

# Display unique 'Main Meal' values and their average ratings
print("Unique Main Meals with Average Ratings:")
print(unique_main_meal_ratings)
```

```python

# Select all columns containing ratings for 'Main Meal'
rating_columns = df.filter(like='Rating Of  Selected Items of  Main Meal [Selected Food]').columns

# Create a new DataFrame with 'Main Meal' and the selected rating columns
selected_data = df[['Main Meal'] + list(rating_columns)].copy()

# Convert the ratings columns to numeric
selected_data[rating_columns] = selected_data[rating_columns].apply(pd.to_numeric, errors='coerce')

# Create a dictionary to store unique values for each item in 'Main Meal'
unique_main_meal_values = {}

# Iterate through each 'Main Meal' and extract unique values for each item
for main_meal in selected_data['Main Meal'].unique():
    item_values = selected_data[selected_data['Main Meal'] == main_meal][rating_columns].values.flatten().tolist()
    unique_main_meal_values[main_meal] = list(set(item_values))

# Display unique values for each item in 'Main Meal'
print("Unique Values for Each Item in 'Main Meal':")
for main_meal, values in unique_main_meal_values.items():
    print(f"{main_meal}: {values}")
```

```python
# Iterate through each 'Main Meal' and extract unique values and average ratings for each item
for main_meal in selected_data['Main Meal'].unique():
    item_data = selected_data[selected_data['Main Meal'] == main_meal]
    unique_values = item_data[rating_columns].values.flatten().tolist()
    unique_main_meal_values[main_meal] = {
        'Unique_Values': list(set(unique_values)),
        'Average_Ratings': item_data[rating_columns].mean().tolist()
    }

# Display unique values and average ratings for each item in 'Main Meal'
print("Unique Values and Average Ratings for Each Item in 'Main Meal':")
for main_meal, values_ratings in unique_main_meal_values.items():
    print(f"{main_meal}:\nUnique Values: {values_ratings['Unique_Values']}\nAverage Ratings: {values_ratings['Average_Ratings']}\n")
```
